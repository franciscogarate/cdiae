# Regresion lineal con Sci-kit

Importo las funciones y librerías que voy a utilizar

```python
from sklearn import linear_model
import matplotlib.pyplot as plt
import pandas as pd
```

Los datos deben ir en columnas

```python
x_train = pd.DataFrame({'x': [1., 2., 3., 4., 5.]})
y_train = pd.DataFrame({'y': [1.25, 5, 11.25, 20, 30.5]})
```

Creo el modelo y lo entreno con datos de entrenamiento

```python
model = linear_model.LinearRegression()
model.fit(x_train, y_train)
```

Hago mi predicción utilizando datos de test (en este ejemplo, para simplificar lo realizo con los datos de entrenamiento)

```python
y_pred = model.predict(x_train)
y_pred
```

Grafico la recta de regresión:

```python
plt.scatter(x_train, y_train)
plt.plot(x_train, y_pred, 'r')
plt.show()
```

Al ser una recta, puedo mostrar la pendiente y la intersección:

```python
model.coef_
```

```python
model.intercept_
```
